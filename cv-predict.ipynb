```python
#IMPORT MODULES
from ultralytics import YOLO
from PIL import Image
import os as os
```

```python
#SELECT PRETRAINED MODEL IN CURRENT DIRECTORY
model = YOLO("./best.pt")
```

```python
#LOAD TEST IMAGES
test_images_dir = "./test_images"

source = [test_images_dir + "/" + s for s in os.listdir(test_images_dir)]
```

```python
#PREDICT
results = model(source)
```

```python
#USE VERBOSE TO INFER DETECTION OF ORANGUTAN
def analyze_results():
    for i, r in enumerate(results):
        detection_string = r.verbose()
        if "no detections" in detection_string:
            print(r.path, "doesn't contain orangutan")
        else:
            print(r.path, "contains orang utan")
```

```python
#PRINTING THE RESULTS
analyze_results()
```
